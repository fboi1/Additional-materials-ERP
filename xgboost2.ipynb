# Fit the XGBoost model

```python
import pandas as pd
import numpy as np
from glob import glob
import os
import joblib
import xgboost as xgb
from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold, TimeSeriesSplit
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

def get_xgboost_param_grid():
    """
    Define hyperparameter grid for XGBoost optimization.
    """
    param_grid = {
        'n_estimators': [100, 200, 300, 500, 600, 800],
        'max_depth': [3, 4, 5, 6, 7, 8],
        'learning_rate': [0.01, 0.05, 0.1, 0.2],
        'subsample': [0.8, 0.9, 1.0],
        'colsample_bytree': [0.8, 0.9, 1.0],
        'reg_alpha': [0, 0.1, 0.5, 1.0],
        'reg_lambda': [0, 0.1, 0.5, 1.0],
        'min_child_weight': [1, 3, 5],
        'gamma': [0, 0.1, 0.2]
    }
    return param_grid

def plot_feature_importance(model, feature_names, stock_name, save_path=None, top_n=20):
    """
    Plot feature importance from XGBoost model.
    """
    # Get feature importance
    importance_df = pd.DataFrame({
        'feature': feature_names,
        'importance': model.feature_importances_
    }).sort_values('importance', ascending=False)
    
    # Plot top N features
    plt.figure(figsize=(12, 8))
    top_features = importance_df.head(top_n)
    
    sns.barplot(data=top_features, x='importance', y='feature', palette='viridis')
    plt.title(f'{stock_name} - Top {top_n} Feature Importance (XGBoost)')
    plt.xlabel('Feature Importance')
    plt.ylabel('Features')
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"Feature importance plot saved to {save_path}")
    
    plt.show()
    
    return importance_df

def plot_confusion_matrix(y_true, y_pred, stock_name, save_path=None):
    """
    Plot confusion matrix.
    """
    cm = confusion_matrix(y_true, y_pred)
    
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=['Down', 'Up'], yticklabels=['Down', 'Up'])
    plt.title(f'{stock_name} - Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"Confusion matrix saved to {save_path}")
    
    plt.show()

# Paths
input_folder = "data/processed/stock_data"
summary_folder = "summary"
model_folder = os.path.join(summary_folder, "xgboost2")
plots_folder = os.path.join(summary_folder, "xgboost_plots2")
os.makedirs(model_folder, exist_ok=True)
os.makedirs(plots_folder, exist_ok=True)

summary_results = []
feature_importance_results = []

# XGBoost hyperparameter search configuration
RANDOM_SEARCH_ITERATIONS = 50  # Reduce for faster execution
CV_FOLDS = 5
RANDOM_STATE = 42

for file_path in glob(os.path.join(input_folder, "*.csv")):
    stock_name = os.path.basename(file_path).replace("_features.csv", "")
    print(f"\n{'='*60}")
    print(f"Processing {stock_name}")
    print(f"{'='*60}")

    try:
        # Load data
        df = pd.read_csv(file_path)
        df.dropna(inplace=True)

        # Create binary target: 1 if price goes up tomorrow, 0 if down
        df['Direction'] = (df['Close'].shift(-1) > df['Close']).astype(int)
        df.dropna(inplace=True)

        # Features
        X = df.drop(columns=[
            "Date","Close","Target","Direction",
            "High","Low","Open","High_lag1","Low_lag1","Open_lag1"
        ])
        y = df['Direction']

        print(f"Dataset shape: {X.shape}")
        print(f"Class distribution: {y.value_counts().to_dict()}")
        print(f"Class balance: {y.mean():.3f} positive class ratio")

        # Split data (time series aware)
        train_size = int(len(X) * 0.8)
        X_train, X_test = X.iloc[:train_size], X.iloc[train_size:]
        y_train, y_test = y.iloc[:train_size], y.iloc[train_size:]

        print(f"Train size: {len(X_train)}, Test size: {len(X_test)}")

        # Optional: Scale features (XGBoost doesn't strictly need it, but can help)
        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train)
        X_test_scaled = scaler.transform(X_test)

        # Convert to DMatrix for XGBoost (more efficient)
        dtrain = xgb.DMatrix(X_train_scaled, label=y_train, feature_names=X.columns.tolist())
        dtest = xgb.DMatrix(X_test_scaled, label=y_test, feature_names=X.columns.tolist())

        # Base XGBoost model
        base_xgb = xgb.XGBClassifier(
            objective='binary:logistic',
            eval_metric='auc',
            random_state=RANDOM_STATE,
            n_jobs=-1,
            verbosity=0
        )

        # Hyperparameter tuning with RandomizedSearchCV
        print(f"Starting hyperparameter tuning with {RANDOM_SEARCH_ITERATIONS} iterations...")
        
        # Use TimeSeriesSplit for time series data
        tscv = TimeSeriesSplit(n_splits=CV_FOLDS)
        
        param_grid = get_xgboost_param_grid()
        
        random_search = RandomizedSearchCV(
            estimator=base_xgb,
            param_distributions=param_grid,
            n_iter=RANDOM_SEARCH_ITERATIONS,
            cv=tscv,
            scoring='roc_auc',
            n_jobs=-1,
            random_state=RANDOM_STATE,
            verbose=0
        )

        # Fit the random search
        random_search.fit(X_train_scaled, y_train)
        
        # Get best model
        best_xgb = random_search.best_estimator_
        
        print(f"Best parameters found:")
        for param, value in random_search.best_params_.items():
            print(f"  {param}: {value}")
        print(f"Best CV AUC: {random_search.best_score_:.4f}")

        # Make predictions
        y_pred = best_xgb.predict(X_test_scaled)
        y_proba = best_xgb.predict_proba(X_test_scaled)[:, 1]

        # Calculate metrics
        acc = accuracy_score(y_test, y_pred)
        auc = roc_auc_score(y_test, y_proba)
        
        print(f"\n--- MODEL PERFORMANCE ---")
        print(f"XGBoost Accuracy: {acc:.4f}")
        print(f"XGBoost AUC: {auc:.4f}")
        print(f"Best CV AUC: {random_search.best_score_:.4f}")
        print("\nClassification Report:")
        print(classification_report(y_test, y_pred))

        # Feature importance analysis
        feature_importance_df = plot_feature_importance(
            best_xgb, X.columns.tolist(), stock_name, 
            save_path=os.path.join(plots_folder, f"{stock_name}_feature_importance.png")
        )
        
        # Plot confusion matrix
        plot_confusion_matrix(
            y_test, y_pred, stock_name,
            save_path=os.path.join(plots_folder, f"{stock_name}_confusion_matrix.png")
        )

        # Save model and scaler
        model_path = os.path.join(model_folder, f"{stock_name}_xgboost_model.pkl")
        scaler_path = os.path.join(model_folder, f"{stock_name}_scaler.pkl")
        params_path = os.path.join(model_folder, f"{stock_name}_best_params.pkl")
        
        joblib.dump(best_xgb, model_path)
        joblib.dump(scaler, scaler_path)
        joblib.dump(random_search.best_params_, params_path)
        
        print(f"Saved model to {model_path}")
        print(f"Saved scaler to {scaler_path}")
        print(f"Saved best parameters to {params_path}")

        # Store top 10 most important features for this stock
        top_features = feature_importance_df.head(10)
        for idx, row in top_features.iterrows():
            feature_importance_results.append({
                'Stock': stock_name,
                'Feature': row['feature'],
                'Importance': row['importance'],
                'Rank': idx + 1
            })

        # Append metrics to summary
        summary_results.append({
            "Stock": stock_name,
            "Accuracy": acc,
            "AUC": auc,
            "CV_AUC": random_search.best_score_,
            "N_Features": X.shape[1],
            "Train_Size": len(X_train),
            "Test_Size": len(X_test),
            "Class_Balance": y.mean(),
            "Best_N_Estimators": random_search.best_params_.get('n_estimators', 'N/A'),
            "Best_Max_Depth": random_search.best_params_.get('max_depth', 'N/A'),
            "Best_Learning_Rate": random_search.best_params_.get('learning_rate', 'N/A')
        })

    except Exception as e:
        print(f"Error processing {stock_name}: {str(e)}")
        continue

# Save summary results
summary_df = pd.DataFrame(summary_results)
summary_file = os.path.join(summary_folder, "stock_data_model_summary_xgboost.csv")
summary_df.to_csv(summary_file, index=False)
print(f"\nSummary saved to {summary_file}")

# Save feature importance results
feature_importance_df = pd.DataFrame(feature_importance_results)
feature_importance_file = os.path.join(summary_folder, "stock_data_feature_importance_xgboost.csv")
feature_importance_df.to_csv(feature_importance_file, index=False)
print(f"Feature importance results saved to {feature_importance_file}")

# Print overall summary
print(f"\n{'='*80}")
print("OVERALL XGBOOST MODEL SUMMARY")
print(f"{'='*80}")
print(f"Total stocks processed: {len(summary_results)}")
print(f"Average Accuracy: {summary_df['Accuracy'].mean():.4f}  {summary_df['Accuracy'].std():.4f}")
print(f"Average AUC: {summary_df['AUC'].mean():.4f}  {summary_df['AUC'].std():.4f}")
print(f"Average CV AUC: {summary_df['CV_AUC'].mean():.4f}  {summary_df['CV_AUC'].std():.4f}")
print(f"Best performing stock (AUC): {summary_df.loc[summary_df['AUC'].idxmax(), 'Stock']} ({summary_df['AUC'].max():.4f})")
print(f"Models with AUC > 0.55: {sum(summary_df['AUC'] > 0.55)}/{len(summary_results)}")

# Most common important features across all stocks
print(f"\nMOST IMPORTANT FEATURES ACROSS ALL STOCKS:")
top_features_overall = feature_importance_df.groupby('Feature')['Importance'].agg(['count', 'mean', 'std']).sort_values('mean', ascending=False)
print(top_features_overall.head(10))

# Create summary plots
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.hist(summary_df['Accuracy'], bins=15, alpha=0.7, color='skyblue', edgecolor='black')
plt.axvline(summary_df['Accuracy'].mean(), color='red', linestyle='--', label=f'Mean: {summary_df["Accuracy"].mean():.3f}')
plt.xlabel('Accuracy')
plt.ylabel('Frequency')
plt.title('Distribution of Model Accuracies')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 3, 2)
plt.hist(summary_df['AUC'], bins=15, alpha=0.7, color='lightcoral', edgecolor='black')
plt.axvline(summary_df['AUC'].mean(), color='red', linestyle='--', label=f'Mean: {summary_df["AUC"].mean():.3f}')
plt.xlabel('AUC')
plt.ylabel('Frequency')
plt.title('Distribution of Model AUCs')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 3, 3)
plt.scatter(summary_df['Accuracy'], summary_df['AUC'], alpha=0.6, s=50)
plt.xlabel('Accuracy')
plt.ylabel('AUC')
plt.title('Accuracy vs AUC')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig(os.path.join(plots_folder, 'overall_model_performance.png'), dpi=300, bbox_inches='tight')
plt.show()

print(f"\nAll plots saved to {plots_folder}")
print("XGBoost model training completed!")
```

# Test the XGBoost model

```python
import pandas as pd
import numpy as np
from glob import glob
import os
import joblib
import xgboost as xgb
from sklearn.model_selection import TimeSeriesSplit
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, roc_auc_score, log_loss, precision_score, recall_score, f1_score
from sklearn.utils import shuffle
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

def calculate_donchian_breakout_metrics(df, predictions, period=20):
    """
    Calculate Donchian breakout strategy metrics including profit factor and Sortino ratio.
    
    Parameters:
    -----------
    df : pd.DataFrame
        DataFrame with OHLC data
    predictions : array-like
        Model predictions (1 for long, 0 for short/hold)
    period : int
        Donchian channel period
    
    Returns:
    --------
    dict : Dictionary containing strategy metrics
    """
    # Create a copy to avoid modifying original
    strategy_df = df.copy()
    
    # Calculate Donchian channels
    strategy_df['Donchian_High'] = strategy_df['High'].rolling(window=period).max()
    strategy_df['Donchian_Low'] = strategy_df['Low'].rolling(window=period).min()
    strategy_df['Donchian_Mid'] = (strategy_df['Donchian_High'] + strategy_df['Donchian_Low']) / 2
    
    # Generate trading signals based on Donchian breakout
    strategy_df['Price'] = strategy_df['Close']
    strategy_df['Signal'] = 0  # 0 = hold, 1 = long, -1 = short
    
    # Donchian breakout signals
    strategy_df.loc[strategy_df['Price'] > strategy_df['Donchian_High'].shift(1), 'Donchian_Signal'] = 1
    strategy_df.loc[strategy_df['Price'] < strategy_df['Donchian_Low'].shift(1), 'Donchian_Signal'] = -1
    strategy_df['Donchian_Signal'].fillna(method='ffill', inplace=True)
    strategy_df['Donchian_Signal'].fillna(0, inplace=True)
    
    # Combine ML predictions with Donchian signals
    # Use ML prediction to filter Donchian signals
    strategy_df['ML_Prediction'] = predictions
    strategy_df['Combined_Signal'] = 0
    
    # Long when both ML predicts up (1) and Donchian is bullish (1)
    strategy_df.loc[(strategy_df['ML_Prediction'] == 1) & 
                    (strategy_df['Donchian_Signal'] == 1), 'Combined_Signal'] = 1
    
    # Short when ML predicts down (0) and Donchian is bearish (-1)
    strategy_df.loc[(strategy_df['ML_Prediction'] == 0) & 
                    (strategy_df['Donchian_Signal'] == -1), 'Combined_Signal'] = -1
    
    # Calculate returns
    strategy_df['Price_Return'] = strategy_df['Price'].pct_change()
    strategy_df['Strategy_Return'] = strategy_df['Combined_Signal'].shift(1) * strategy_df['Price_Return']
    
    # Remove NaN values
    valid_mask = ~(strategy_df['Strategy_Return'].isna() | strategy_df['Price_Return'].isna())
    returns = strategy_df.loc[valid_mask, 'Strategy_Return']
    price_returns = strategy_df.loc[valid_mask, 'Price_Return']
    
    if len(returns) == 0:
        return {
            'total_return': 0,
            'annualized_return': 0,
            'volatility': 0,
            'sharpe_ratio': 0,
            'sortino_ratio': 0,
            'max_drawdown': 0,
            'profit_factor': 1,
            'win_rate': 0,
            'avg_win': 0,
            'avg_loss': 0,
            'total_trades': 0
        }
    
    # Calculate basic metrics
    total_return = (1 + returns).prod() - 1
    annualized_return = (1 + total_return) ** (252 / len(returns)) - 1
    volatility = returns.std() * np.sqrt(252)
    
    # Sharpe ratio
    risk_free_rate = 0.02  # Assume 2% risk-free rate
    excess_returns = returns - risk_free_rate / 252
    sharpe_ratio = excess_returns.mean() / returns.std() * np.sqrt(252) if returns.std() > 0 else 0
    
    # Sortino ratio
    negative_returns = returns[returns < 0]
    downside_deviation = negative_returns.std() * np.sqrt(252) if len(negative_returns) > 0 else 0.001
    sortino_ratio = (annualized_return - risk_free_rate) / downside_deviation if downside_deviation > 0 else 0
    
    # Maximum drawdown
    cumulative_returns = (1 + returns).cumprod()
    rolling_max = cumulative_returns.expanding().max()
    drawdown = (cumulative_returns - rolling_max) / rolling_max
    max_drawdown = drawdown.min()
    
    # Profit factor and win rate
    winning_trades = returns[returns > 0]
    losing_trades = returns[returns < 0]
    
    total_wins = winning_trades.sum()
    total_losses = abs(losing_trades.sum())
    profit_factor = total_wins / total_losses if total_losses > 0 else float('inf')
    
    win_rate = len(winning_trades) / len(returns) if len(returns) > 0 else 0
    avg_win = winning_trades.mean() if len(winning_trades) > 0 else 0
    avg_loss = losing_trades.mean() if len(losing_trades) > 0 else 0
    
    # Count actual trades (signal changes)
    signals = strategy_df['Combined_Signal'].fillna(0)
    total_trades = (signals.diff() != 0).sum()
    
    return {
        'total_return': total_return,
        'annualized_return': annualized_return,
        'volatility': volatility,
        'sharpe_ratio': sharpe_ratio,
        'sortino_ratio': sortino_ratio,
        'max_drawdown': max_drawdown,
        'profit_factor': profit_factor,
        'win_rate': win_rate,
        'avg_win': avg_win,
        'avg_loss': avg_loss,
        'total_trades': total_trades,
        'benchmark_return': (1 + price_returns).prod() - 1  # Buy and hold return
    }

def monte_carlo_permutation_test_xgboost(X_train, X_test, y_train, y_test, 
                                        df_test, best_params=None, n_permutations=1000, 
                                        random_state=42, use_cv=True, donchian_period=20):
    """
    Perform Monte Carlo permutation test for XGBoost model with Donchian breakout strategy.
    
    Parameters:
    -----------
    X_train, X_test : array-like
        Training and testing features
    y_train, y_test : array-like
        Training and testing labels
    df_test : pd.DataFrame
        Test DataFrame with OHLC data for strategy calculation
    best_params : dict
        Best hyperparameters found from tuning
    n_permutations : int
        Number of permutations for the test
    random_state : int
        Random seed for reproducibility
    use_cv : bool
        Whether to use cross-validation for more robust estimates
    donchian_period : int
        Period for Donchian channel calculation
    
    Returns:
    --------
    dict : Dictionary containing test results
    """
    np.random.seed(random_state)
    
    # Scale features
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    # Default XGBoost parameters if none provided
    if best_params is None:
        best_params = {
            'n_estimators': 200,
            'max_depth': 5,
            'learning_rate': 0.1,
            'subsample': 0.9,
            'colsample_bytree': 0.9,
            'reg_alpha': 0.1,
            'reg_lambda': 0.1,
            'min_child_weight': 3,
            'gamma': 0.1
        }
    
    # Create base model with best parameters
    base_model_params = {
        'objective': 'binary:logistic',
        'eval_metric': 'auc',
        'random_state': random_state,
        'n_jobs': -1,
        'verbosity': 0,
        **best_params
    }
    
    # Fit original model and get baseline metrics
    original_model = xgb.XGBClassifier(**base_model_params)
    original_model.fit(X_train_scaled, y_train)
    
    # Original predictions and metrics
    y_pred_original = original_model.predict(X_test_scaled)
    y_proba_original = original_model.predict_proba(X_test_scaled)[:, 1]
    
    # Calculate ML metrics for original model
    original_ml_metrics = {
        'accuracy': accuracy_score(y_test, y_pred_original),
        'auc': roc_auc_score(y_test, y_proba_original),
        'precision': precision_score(y_test, y_pred_original, zero_division=0),
        'recall': recall_score(y_test, y_pred_original, zero_division=0),
        'f1': f1_score(y_test, y_pred_original, zero_division=0),
        'log_loss': log_loss(y_test, y_proba_original)
    }
    
    # Calculate Donchian strategy metrics for original model
    original_strategy_metrics = calculate_donchian_breakout_metrics(
        df_test, y_pred_original, donchian_period
    )
    
    # Combine all original metrics
    original_metrics = {**original_ml_metrics, **original_strategy_metrics}
    
    # Cross-validation score for original model (if requested)
    if use_cv:
        tscv = TimeSeriesSplit(n_splits=3)  # Reduced splits for speed
        cv_scores = []
        for train_idx, val_idx in tscv.split(X_train_scaled):
            X_cv_train, X_cv_val = X_train_scaled[train_idx], X_train_scaled[val_idx]
            y_cv_train, y_cv_val = y_train.iloc[train_idx], y_train.iloc[val_idx]
            
            cv_model = xgb.XGBClassifier(**base_model_params)
            cv_model.fit(X_cv_train, y_cv_train)
            cv_proba = cv_model.predict_proba(X_cv_val)[:, 1]
            cv_scores.append(roc_auc_score(y_cv_val, cv_proba))
        
        original_metrics['cv_auc'] = np.mean(cv_scores)
        original_metrics['cv_auc_std'] = np.std(cv_scores)
    
    # Initialize arrays to store permutation results
    permuted_metrics = {metric: [] for metric in original_metrics.keys() if not metric.endswith('_std')}
    
    print(f"Running {n_permutations} permutations for XGBoost model with Donchian strategy...")
    
    # Perform permutation test
    for i in range(n_permutations):
        try:
            # Shuffle the training labels
            y_train_permuted = shuffle(y_train, random_state=random_state + i)
            
            # Train model on permuted data
            permuted_model = xgb.XGBClassifier(**base_model_params)
            permuted_model.fit(X_train_scaled, y_train_permuted)
            
            # Make predictions on original test set
            y_pred_permuted = permuted_model.predict(X_test_scaled)
            y_proba_permuted = permuted_model.predict_proba(X_test_scaled)[:, 1]
            
            # Calculate ML metrics for permuted model
            permuted_ml_metrics = {
                'accuracy': accuracy_score(y_test, y_pred_permuted),
                'auc': roc_auc_score(y_test, y_proba_permuted),
                'precision': precision_score(y_test, y_pred_permuted, zero_division=0),
                'recall': recall_score(y_test, y_pred_permuted, zero_division=0),
                'f1': f1_score(y_test, y_pred_permuted, zero_division=0),
                'log_loss': log_loss(y_test, y_proba_permuted)
            }
            
            # Calculate Donchian strategy metrics for permuted model
            permuted_strategy_metrics = calculate_donchian_breakout_metrics(
                df_test, y_pred_permuted, donchian_period
            )
            
            # Store all permuted metrics
            all_permuted_metrics = {**permuted_ml_metrics, **permuted_strategy_metrics}
            for metric, value in all_permuted_metrics.items():
                if metric in permuted_metrics:
                    permuted_metrics[metric].append(value)
            
            # Cross-validation for permuted model (if requested and occasionally)
            if use_cv and i % 50 == 0:  # Only every 50th iteration for speed
                cv_scores_perm = []
                for train_idx, val_idx in tscv.split(X_train_scaled):
                    X_cv_train, X_cv_val = X_train_scaled[train_idx], X_train_scaled[val_idx]
                    y_cv_train, y_cv_val = y_train_permuted.iloc[train_idx], y_train_permuted.iloc[val_idx]
                    
                    cv_model_perm = xgb.XGBClassifier(**base_model_params)
                    cv_model_perm.fit(X_cv_train, y_cv_train)
                    cv_proba_perm = cv_model_perm.predict_proba(X_cv_val)[:, 1]
                    cv_scores_perm.append(roc_auc_score(y_cv_val, cv_proba_perm))
                
                if 'cv_auc' in permuted_metrics:
                    permuted_metrics['cv_auc'].append(np.mean(cv_scores_perm))
                else:
                    permuted_metrics['cv_auc'] = [np.mean(cv_scores_perm)]
            
        except Exception as e:
            print(f"Error in permutation {i}: {e}")
            # Add NaN values to maintain array lengths
            for metric in permuted_metrics.keys():
                if metric != 'cv_auc' or len(permuted_metrics[metric]) < i:
                    permuted_metrics[metric].append(np.nan)
    
    # Convert to numpy arrays and remove NaN values
    for metric in permuted_metrics.keys():
        permuted_metrics[metric] = np.array(permuted_metrics[metric])
        permuted_metrics[metric] = permuted_metrics[metric][~np.isnan(permuted_metrics[metric])]
    
    # Calculate p-values and effect sizes
    results = {'original_metrics': original_metrics}
    
    for metric, values in permuted_metrics.items():
        if len(values) > 0:
            original_value = original_metrics[metric]
            
            # P-value (one-tailed test - testing if original is better than random)
            if metric in ['log_loss', 'max_drawdown', 'avg_loss', 'volatility']:  # Lower is better
                p_value = np.mean(values <= original_value)
            else:  # Higher is better for other metrics
                p_value = np.mean(values >= original_value)
            
            # Effect size (standardized difference)
            if np.std(values) > 0:
                effect_size = (original_value - np.mean(values)) / np.std(values)
            else:
                effect_size = 0
            
            results[f'permuted_{metric}'] = values
            results[f'p_value_{metric}'] = p_value
            results[f'effect_size_{metric}'] = effect_size
            results[f'permuted_mean_{metric}'] = np.mean(values)
            results[f'permuted_std_{metric}'] = np.std(values)
    
    return results

def plot_permutation_results_xgboost(results, stock_name, save_path=None):
    """
    Plot the results of XGBoost permutation test with ML and strategy metrics.
    """
    # Select key metrics to plot (mix of ML and strategy metrics)
    metrics_to_plot = ['auc', 'profit_factor', 'sortino_ratio', 'sharpe_ratio']
    available_metrics = [m for m in metrics_to_plot if f'permuted_{m}' in results]
    
    n_metrics = len(available_metrics)
    if n_metrics == 0:
        print(f"No metrics available for plotting for {stock_name}")
        return
    
    # Determine subplot layout
    rows = min(2, (n_metrics + 1) // 2)
    cols = min(2, n_metrics)
    fig, axes = plt.subplots(rows, cols, figsize=(16, 8 * rows))
    
    if n_metrics == 1:
        axes = [axes]
    elif rows == 1:
        axes = axes if isinstance(axes, (list, np.ndarray)) else [axes]
    else:
        axes = axes.flatten()
    
    colors = ['lightblue', 'lightcoral', 'lightgreen', 'lightyellow']
    
    for i, metric in enumerate(available_metrics):
        if i >= len(axes):
            break
            
        ax = axes[i]
        
        # Plot histogram of permuted results
        permuted_values = results[f'permuted_{metric}']
        original_value = results['original_metrics'][metric]
        p_value = results[f'p_value_{metric}']
        effect_size = results[f'effect_size_{metric}']
        
        # Handle infinite values
        if np.isinf(original_value):
            original_value = max(permuted_values) * 2 if len(permuted_values) > 0 else 1
        permuted_values = permuted_values[~np.isinf(permuted_values)]
        
        if len(permuted_values) > 0:
            ax.hist(permuted_values, bins=50, alpha=0.7, color=colors[i % len(colors)], 
                    label='Permuted Distribution', density=True)
            ax.axvline(original_value, color='red', linestyle='--', linewidth=2, 
                       label=f'Original: {original_value:.4f}')
            ax.axvline(np.mean(permuted_values), color='green', linestyle='-', linewidth=2,
                       label=f'Permuted Mean: {np.mean(permuted_values):.4f}')
        
        ax.set_xlabel(metric.replace('_', ' ').title())
        ax.set_ylabel('Density')
        ax.set_title(f'{stock_name} - {metric.replace("_", " ").title()} Permutation Test\n'
                     f'P-value: {p_value:.4f}, Effect Size: {effect_size:.2f}')
        ax.legend()
        ax.grid(True, alpha=0.3)
    
    # Hide unused subplots
    for i in range(len(available_metrics), len(axes)):
        axes[i].set_visible(False)
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"XGBoost permutation test plot saved to {save_path}")
    
    plt.show()

def create_significance_summary_plot(permutation_results_df, save_path=None):
    """
    Create summary plots of statistical significance across all stocks.
    """
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    
    # Key metrics to analyze
    metrics = ['auc', 'profit_factor', 'sortino_ratio', 'sharpe_ratio']
    
    for i, metric in enumerate(metrics):
        ax = axes[i//2, i%2]
        p_col = f'P_Value_{metric.upper()}'
        
        if p_col in permutation_results_df.columns:
            p_values = permutation_results_df[p_col].dropna()
            
            ax.hist(p_values, bins=20, alpha=0.7, edgecolor='black')
            ax.axvline(0.05, color='red', linestyle='--', linewidth=2, label=' = 0.05')
            ax.axvline(0.01, color='orange', linestyle='--', linewidth=2, label=' = 0.01')
            
            significant_05 = sum(p_values < 0.05)
            significant_01 = sum(p_values < 0.01)
            
            ax.set_xlabel('P-value')
            ax.set_ylabel('Frequency')
            ax.set_title(f'{metric.replace("_", " ").title()} P-values Distribution\n'
                        f'Significant (p<0.05): {significant_05}/{len(p_values)} stocks\n'
                        f'Highly significant (p<0.01): {significant_01}/{len(p_values)} stocks')
            ax.legend()
            ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"Significance summary plot saved to {save_path}")
    
    plt.show()

# Main execution
if __name__ == "__main__":
    # Configuration
    N_PERMUTATIONS = 500  # Reduced for XGBoost (slower than logistic regression)
    RANDOM_STATE = 42
    USE_CV = True  # Set to False for faster execution
    DONCHIAN_PERIOD = 20  # 20-day Donchian channel
    
    # Paths
    input_folder = "data/processed/stock_data"
    summary_folder = "summary"
    model_folder = os.path.join(summary_folder, "xgboost2")
    permutation_folder = os.path.join(summary_folder, "xgboost_donchian_permutation_tests")
    os.makedirs(permutation_folder, exist_ok=True)
    
    summary_results = []
    permutation_results = []
    
    # Process each stock file
    for file_path in glob(os.path.join(input_folder, "*.csv")):
        stock_name = os.path.basename(file_path).replace("_features.csv", "")
        print(f"\n{'='*60}")
        print(f"Processing {stock_name}")
        print(f"{'='*60}")
        
        try:
            # Load data
            df = pd.read_csv(file_path)
            df = df.tail(1000)
            df.dropna(inplace=True)
            
            # Create binary target: 1 if price goes up tomorrow, 0 if down
            df['Direction'] = (df['Close'].shift(-1) > df['Close']).astype(int)
            df.dropna(inplace=True)
            
            # Features
            X = df.drop(columns=[
                "Date","Close","Target","Direction",
                "High","Low","Open","High_lag1","Low_lag1","Open_lag1"
            ])
            y = df['Direction']
            
            # Split
            train_size = int(len(X) * 0.8)
            X_train, X_test = X.iloc[:train_size], X.iloc[train_size:]
            y_train, y_test = y.iloc[:train_size], y.iloc[train_size:]
            df_test = df.iloc[train_size:]  # For strategy calculation
            
            print(f"Train size: {len(X_train)}, Test size: {len(X_test)}")
            print(f"Features: {X.shape[1]}")
            print(f"Class distribution - Train: {y_train.value_counts().to_dict()}")
            print(f"Class distribution - Test: {y_test.value_counts().to_dict()}")
            
            # Load best parameters if available
            params_path = os.path.join(model_folder, f"{stock_name}_best_params.pkl")
            if os.path.exists(params_path):
                best_params = joblib.load(params_path)
                print(f"Loaded best parameters from {params_path}")
            else:
                best_params = None
                print("Using default parameters")
            
            # Perform Monte Carlo permutation test with Donchian strategy
            results = monte_carlo_permutation_test_xgboost(
                X_train, X_test, y_train, y_test, df_test,
                best_params=best_params,
                n_permutations=N_PERMUTATIONS,
                random_state=RANDOM_STATE,
                use_cv=USE_CV,
                donchian_period=DONCHIAN_PERIOD
            )
            
            # Print results
            print(f"\n--- XGBOOST + DONCHIAN STRATEGY PERMUTATION TEST RESULTS ---")
            original_metrics = results['original_metrics']
            
            # Print ML metrics
            ml_metrics = ['accuracy', 'auc', 'precision', 'recall', 'f1']
            print("\nML METRICS:")
            for metric in ml_metrics:
                if metric in original_metrics:
                    value = original_metrics[metric]
                    print(f"Original {metric.upper()}: {value:.4f}")
                    
                    if f'p_value_{metric}' in results:
                        p_val = results[f'p_value_{metric}']
                        effect_size = results[f'effect_size_{metric}']
                        print(f"  P-value: {p_val:.4f}, Effect Size: {effect_size:.4f}")
            
            # Print strategy metrics
            strategy_metrics = ['total_return', 'annualized_return', 'sharpe_ratio', 'sortino_ratio', 
                              'profit_factor', 'max_drawdown', 'win_rate']
            print("\nSTRATEGY METRICS:")
            for metric in strategy_metrics:
                if metric in original_metrics:
                    value = original_metrics[metric]
                    if metric in ['total_return', 'annualized_return', 'max_drawdown']:
                        print(f"Original {metric.upper()}: {value:.2%}")
                    elif metric == 'win_rate':
                        print(f"Original {metric.upper()}: {value:.2%}")
                    else:
                        print(f"Original {metric.upper()}: {value:.4f}")
                    
                    if f'p_value_{metric}' in results:
                        p_val = results[f'p_value_{metric}']
                        effect_size = results[f'effect_size_{metric}']
                        print(f"  P-value: {p_val:.4f}, Effect Size: {effect_size:.4f}")
            
            print(f"\nBenchmark (Buy & Hold) Return: {original_metrics.get('benchmark_return', 0):.2%}")
            print(f"Total Trades: {original_metrics.get('total_trades', 0)}")
            
            # Create and save plots
            plot_path = os.path.join(permutation_folder, f"{stock_name}_donchian_permutation_test.png")
            plot_permutation_results_xgboost(results, stock_name, plot_path)
            
            # Store results for summary
            result_dict = {
                "Stock": stock_name,
                "N_Permutations": N_PERMUTATIONS,
                "Train_Size": len(X_train),
                "Test_Size": len(X_test),
                "N_Features": X.shape[1],
                "Donchian_Period": DONCHIAN_PERIOD
            }
            
            # Add original metrics and statistical results
            for metric, value in original_metrics.items():
                if not metric.endswith('_std'):
                    result_dict[f'Original_{metric.upper()}'] = value
                    
                    if f'p_value_{metric}' in results:
                        result_dict[f'P_Value_{metric.upper()}'] = results[f'p_value_{metric}']
                        result_dict[f'Effect_Size_{metric.upper()}'] = results[f'effect_size_{metric}']
                        result_dict[f'Permuted_Mean_{metric.upper()}'] = results[f'permuted_mean_{metric}']
                        result_dict[f'Permuted_Std_{metric.upper()}'] = results[f'permuted_std_{metric}']
                        result_dict[f'Significant_{metric.upper()}'] = results[f'p_value_{metric}'] < 0.05
            
            permutation_results.append(result_dict)
            
            # Store summary for compatibility
            summary_results.append({
                "Stock": stock_name,
                "Accuracy": original_metrics['accuracy'],
                "AUC": original_metrics['auc'],
                "F1": original_metrics['f1'],
                "Precision": original_metrics['precision'],
                "Total_Return": original_metrics['total_return'],
                "Sortino_Ratio": original_metrics['sortino_ratio'],
                "Profit_Factor": original_metrics['profit_factor']
            })
            
        except Exception as e:
            print(f"Error processing {stock_name}: {str(e)}")
            continue
    
    # Save detailed permutation test results
    permutation_df = pd.DataFrame(permutation_results)
    permutation_file = os.path.join(summary_folder, "xgboost_donchian_permutation_test_results.csv")
    permutation_df.to_csv(permutation_file, index=False)
    print(f"\nXGBoost Donchian permutation test results saved to {permutation_file}")
    
    # Save summary results
    summary_df = pd.DataFrame(summary_results)
    summary_file = os.path.join(summary_folder, "xgboost_donchian_model_summary.csv")
    summary_df.to_csv(summary_file, index=False)
    print(f"Summary saved to {summary_file}")
    
    # Create significance summary plots
    significance_plot_path = os.path.join(permutation_folder, "significance_summary_donchian.png")
    create_significance_summary_plot(permutation_df, significance_plot_path)
    
    # Print comprehensive summary
    print(f"\n{'='*80}")
    print("COMPREHENSIVE XGBOOST + DONCHIAN STRATEGY PERMUTATION TEST SUMMARY")
    print(f"{'='*80}")
    print(f"Total stocks processed: {len(permutation_results)}")
    print(f"Number of permutations per stock: {N_PERMUTATIONS}")
    print(f"Donchian channel period: {DONCHIAN_PERIOD} days")
    
    # Significance summary for ML metrics
    print(f"\n--- ML METRICS SIGNIFICANCE ---")
    ml_metrics = ['ACCURACY', 'AUC', 'F1', 'PRECISION']
    for metric in ml_metrics:
        p_col = f'P_Value_{metric}'
        if p_col in permutation_df.columns:
            significant_05 = sum(permutation_df[p_col] < 0.05)
            significant_01 = sum(permutation_df[p_col] < 0.01)
            mean_p_value = permutation_df[p_col].mean()
            mean_effect_size = permutation_df[f'Effect_Size_{metric}'].mean() if f'Effect_Size_{metric}' in permutation_df.columns else 0
            
            print(f"\n{metric}:")
            print(f"  Significant stocks (p < 0.05): {significant_05}/{len(permutation_results)} ({100*significant_05/len(permutation_results):.1f}%)")
            print(f"  Highly significant stocks (p < 0.01): {significant_01}/{len(permutation_results)} ({100*significant_01/len(permutation_results):.1f}%)")
            print(f"  Mean p-value: {mean_p_value:.4f}")
            print(f"  Mean effect size: {mean_effect_size:.4f}")
    
    # Significance summary for strategy metrics
    print(f"\n--- STRATEGY METRICS SIGNIFICANCE ---")
    strategy_metrics = ['PROFIT_FACTOR', 'SORTINO_RATIO', 'SHARPE_RATIO', 'TOTAL_RETURN']
    for metric in strategy_metrics:
        p_col = f'P_Value_{metric}'
        if p_col in permutation_df.columns:
            significant_05 = sum(permutation_df[p_col] < 0.05)
            significant_01 = sum(permutation_df[p_col] < 0.01)
            mean_p_value = permutation_df[p_col].mean()
            mean_effect_size = permutation_df[f'Effect_Size_{metric}'].mean() if f'Effect_Size_{metric}' in permutation_df.columns else 0
            
            print(f"\n{metric}:")
            print(f"  Significant stocks (p < 0.05): {significant_05}/{len(permutation_results)} ({100*significant_05/len(permutation_results):.1f}%)")
            print(f"  Highly significant stocks (p < 0.01): {significant_01}/{len(permutation_results)} ({100*significant_01/len(permutation_results):.1f}%)")
            print(f"  Mean p-value: {mean_p_value:.4f}")
            print(f"  Mean effect size: {mean_effect_size:.4f}")
    
    # Top performing stocks by different metrics
    print(f"\n--- TOP 5 PERFORMING STOCKS ---")
    
    if 'P_Value_PROFIT_FACTOR' in permutation_df.columns:
        print(f"\nTOP 5 STOCKS BY PROFIT FACTOR P-VALUE:")
        top_pf = permutation_df.nsmallest(5, 'P_Value_PROFIT_FACTOR')[
            ['Stock', 'Original_PROFIT_FACTOR', 'P_Value_PROFIT_FACTOR', 'Effect_Size_PROFIT_FACTOR']
        ]
        print(top_pf.to_string(index=False))
    
    if 'P_Value_SORTINO_RATIO' in permutation_df.columns:
        print(f"\nTOP 5 STOCKS BY SORTINO RATIO P-VALUE:")
        top_sortino = permutation_df.nsmallest(5, 'P_Value_SORTINO_RATIO')[
            ['Stock', 'Original_SORTINO_RATIO', 'P_Value_SORTINO_RATIO', 'Effect_Size_SORTINO_RATIO']
        ]
        print(top_sortino.to_string(index=False))
    
    if 'P_Value_TOTAL_RETURN' in permutation_df.columns:
        print(f"\nTOP 5 STOCKS BY TOTAL RETURN P-VALUE:")
        top_return = permutation_df.nsmallest(5, 'P_Value_TOTAL_RETURN')[
            ['Stock', 'Original_TOTAL_RETURN', 'P_Value_TOTAL_RETURN', 'Effect_Size_TOTAL_RETURN']
        ]
        print(top_return.to_string(index=False))
    
    # Strategy performance summary
    print(f"\n--- STRATEGY PERFORMANCE SUMMARY ---")
    if 'Original_TOTAL_RETURN' in permutation_df.columns:
        positive_returns = sum(permutation_df['Original_TOTAL_RETURN'] > 0)
        profitable_stocks = sum(permutation_df['Original_PROFIT_FACTOR'] > 1) if 'Original_PROFIT_FACTOR' in permutation_df.columns else 0
        
        print(f"Stocks with positive returns: {positive_returns}/{len(permutation_results)} ({100*positive_returns/len(permutation_results):.1f}%)")
        print(f"Stocks with profit factor > 1: {profitable_stocks}/{len(permutation_results)} ({100*profitable_stocks/len(permutation_results):.1f}%)")
        
        avg_return = permutation_df['Original_TOTAL_RETURN'].mean()
        avg_sortino = permutation_df['Original_SORTINO_RATIO'].mean() if 'Original_SORTINO_RATIO' in permutation_df.columns else 0
        avg_profit_factor = permutation_df['Original_PROFIT_FACTOR'].mean() if 'Original_PROFIT_FACTOR' in permutation_df.columns else 0
        
        print(f"Average total return: {avg_return:.2%}")
        print(f"Average Sortino ratio: {avg_sortino:.4f}")
        print(f"Average profit factor: {avg_profit_factor:.4f}")
    
    print(f"\nXGBoost + Donchian breakout strategy permutation testing completed!")
    print(f"Results saved to: {permutation_file}")
    print(f"Plots saved to: {permutation_folder}")
    print(f"\nKey insights:")
    print(f"- Used {DONCHIAN_PERIOD}-day Donchian channels for breakout signals")
    print(f"- Combined ML predictions with technical analysis")
    print(f"- Added Sortino ratio for downside risk assessment")
    print(f"- Profit factor measures gross profit to gross loss ratio")
```
